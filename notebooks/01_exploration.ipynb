## Dataset Exploration
In this notebook, we explore the resumes dataset to check for gender distribution, selection distribution, and potential bias in hiring.

## Gender Distribution
Visualizing the count of male and female candidates.

## Selection Distribution
Checking how many candidates were selected vs rejected.

## Word Cloud of Selected Candidates
Visualizing most frequent words in resumes of selected candidates.



```python
# 01_exploration.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Load dataset
df = pd.read_csv("data/processed/processed_data.csv")

# Basic info
df.info()
df.head()

# Gender distribution
sns.countplot(x="gender", data=df)
plt.title("Gender Distribution")
plt.show()

# Label distribution
sns.countplot(x="label", data=df)
plt.title("Selection Distribution")
plt.show()

# Word cloud for selected resumes
text = " ".join(df[df["label"]==1]["cleaned_resume"].tolist())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud: Selected Candidates")
plt.show()
```
